```python
# This program is coded to describe the motion of a system with 2 blocks and three springs.
# Here M1=M2, and K1=K12=K2

# Importing the essential libararies
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# initial conditions

t0 = 0      # initial time
dt = 0.01   # step in time
tf = 50     # max time

# Hooke's constant
K1 = 1
K12 = 1
K2 = 1

# Mass for the 2 blocks 
M1 = 1
M2 = 1

# the distance between the blocks
L1 = 10
L12 = 10
L2 = 10

# defining the equation for Hooke's Law for x1 
def d2x1dt2(x1,x2):
    return(-(K1/M1)*(x1-L1)+(K12/M1)*(x2-x1-L12))



# defining the equation for Hooke's Law for x2
def d2x2dt2(x1,x2):
    return((K12/M2)*(L12-x2+x1)+(K2/M2)*(L1+L12-x2))


# defining the equation for the spring function for different intial velovities and displacments
def Masses_springs(V10,V20,x10,x20):
    
    # lists for the position, time, and velocity.
    x1 = [x10]
    x2 = [x20]
    V1 = [V10]
    V2 = [V20]
    t = [t0]
    
    # while loop using Euler equations for the position, time, and velocity
    while t[-1] < tf:
        
        x1.append(x1[-1] + V1[-1]*dt)
        x2.append(x2[-1] + V2[-1]*dt)
        V1.append(V1[-1] + d2x1dt2(x1[-1],x2[-1])*dt)
        V2.append(V2[-1] + d2x2dt2(x1[-1],x2[-1])*dt)
        t.append(t[-1] + dt)
        
    return x1,x2,V1,V2,t

# Test 1: stationary masses.
x11, x21, V11, V21, t1 = Masses_springs(0,0,10,20)
# Test 2: starting with displacement for M1 and M2 
x12, x22, V12, V22, t2 = Masses_springs(0,0,9,21)
# Test 3: starting with intial veloceties
x13, x23, V13, V23, t3 = Masses_springs(3,3,10,20)

# graph of x1(t)
plt.figure(1)
plt.plot( t1, x11, "red")    # x1 for test 1
plt.plot( t2, x12, "blue")   # x1 for test 2
plt.plot( t3, x13, "orange")  # x1 for test 3

# graph of x2(t)
plt.figure(2)
plt.plot( t1, x21, "red")  # x2 for test 1
plt.plot( t2, x22, "blue")   # x2 for test 2
plt.plot( t3, x23, "orange")   # x2 for test 3
```
