```python
# This code is programed to describe the rotation of the earth around the sun. Here, I have used simplified constansts
# for the gravity and the masses of the earth and the sun.

import numpy as np
import matplotlib.pyplot as plt


# Initial conditions

t0 = 0

tf = 10

dt = 0.001

G = 1         # Gravity constant

M = 1         # Mass of the sun

m = 0.0001     # Mass of the earth

x0 = 1

y0 = 0

Vy0 = 1

Vx0 = 0


# differencial equations for the acceleration of the x and y components

def d2xdt2(x,y):
    return(-G*M*x/((x**2+y**2)**1.5))
def d2ydt2(x,y):
    return(-G*M*y/((x**2+y**2)**1.5))

# lists for the time, positions, and velocities


t = [t0]

x = [x0]

y = [y0]

Vx = [Vx0]

Vy = [Vy0]

# a loop using euler equation for the positions and velocities
while t[-1] < tf:
    t.append(t[-1] + dt)
    x.append(x[-1] + Vx[-1]*dt)
    y.append(y[-1] + Vy[-1]*dt)
    Vx.append(Vx[-1] + d2xdt2(x[-1], y[-1])*dt)
    Vy.append(Vy[-1] + d2ydt2(x[-1], y[-1])*dt)




plt.plot(x,y)
plt.title("Orbit of the Earth")
plt.xlabel("x")
plt.ylabel("y")
```
