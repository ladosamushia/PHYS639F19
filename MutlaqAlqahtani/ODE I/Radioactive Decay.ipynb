```python
# this code solves the differencial equation of radioactive decay

import numpy as np
import matplotlib.pyplot as plt



# Establishing a range for the decay
r = np.linspace(0.1, .9, 10)


# Function for the differencial equation of the number of atoms
def dNdt(n, r):
    return(-n/r)

# A loop to construct the euler equation with the intial conditions 
for ri in r:
    n0 = 100000 # number of atoms
    tmax = 10 # max time
    t0 = 0 # intial time
    dt = 0.0001 # step
    
    # lists for time and atoms 
    t = [t0]
    n = [n0]
    
    
    
    # loop to instruct the limits
    
    while t[-1] < tmax:
        n.append(n[-1] + dNdt(n[-1], ri)*dt)
        t.append(t[-1] + dt)
        
    plt.plot(t, n)

    
plt.grid(True)
```
