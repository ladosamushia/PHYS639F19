```python
# this code solves the differencial equation of radioactive decay of two atoms decaying into each other.

import numpy as np
import matplotlib.pyplot as plt



# intial coefs

ra = 0.5
rb = 0.7

# Function for the differencial equation of the number of atoms

def dNadt(na, ra, nb, rb):
    return(-na/ra + nb/rb)
def dNbdt(nb, rb, na, ra):
    return(-nb/rb + na/ra)

# the intial conditions 
    
nb0 = 200000
na0 = 1000000 # number of atoms
tmax = 10 # max time
t0 = 0 # intial time
dt = 0.0001 # step
    
# lists for time and atoms 
t = [t0]
na = [na0]
nb = [nb0]
    

# loop to instruct the limits
    
while t[-1] < tmax:
    na.append(na[-1] + dNadt(na[-1], ra, nb[-1], rb)*dt)
    nb.append(nb[-1] + dNbdt(nb[-1], rb, na[-1], ra)*dt)
    t.append(t[-1] + dt)
        
plt.plot(t, na)
plt.plot(t, nb)
plt.grid(True)
```
