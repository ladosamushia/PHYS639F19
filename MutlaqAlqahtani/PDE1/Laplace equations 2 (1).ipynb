```python
# This program is coded to describe laplace equations of two plates.


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Nsteps = 1000       # number of steps

# creaing the matrix m
m = np.zeros((100,100))

# the plates
m[30:70,30] = 1
m[30:70,70] = -1

# ploting the first matrix as a test to show the plates
plt.figure(1)
plt.imshow(m)     


# creaing another large matrix m 
m = np.zeros((101,101))


# the plates
m[30:70,30] = 1
m[30:70,70] = -1

for s in range(Nsteps):
    for i in range(100):
        for j in range(100):
            
            # calculating the values of each entry in the matrix
            m[i,j] = (1/4)*(m[i+1,j] + m[i-1,j] + m[i,j+1] + m[i,j-1])
            # the bounderies in m
            m[30:70,30] = 1
            m[30:70,70] = -1

# ploting the values
plt.figure(2)
plt.imshow(m)     
```
