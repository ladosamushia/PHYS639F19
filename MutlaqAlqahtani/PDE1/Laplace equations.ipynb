```python
# This program is coded to describe laplace equations of a square inside of another square. 



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


Nsteps = 1000       # number of steps

# creating a matrix m 
m = np.zeros((101,101))

# the boundery on the edges of the big square
m[0,:] = -5
m[:,0] = -5
m[99,:] = -5
m[:,99] = -5


# the boundery on the edges of the small square
m[40,40:60]= 7
m[40:60,40]= 7
m[60,40:60]= 7
m[40:61,60]= 7


# ploting a test for the square inside of another 
plt.figure(1)
plt.imshow(m)     

# creating a loop to count for the laplance equation
for s in range(Nsteps):
    for i in range(100):
        for j in range(100):
            
            # calculating the values of each entry in the matrix
            
            m[i,j] = (1/4)*(m[i+1,j] + m[i-1,j] + m[i,j+1] + m[i,j-1])
            # the bounderies in m
            m[0,:] = -5
            m[:,0] = -5
            m[99,:] = -5
            m[:,99] = -5

            m[40,40:60]= 7
            m[40:60,40]= 7
            m[60,40:60]= 7
            m[40:61,60]= 7

# ploting the values
plt.figure(2)
plt.imshow(m) 
```
