```python
# This program is coded to discribe the motion of a mass on a spring by using Hooke's Law and differencial equations


import numpy as np
import matplotlib.pyplot as plt

# initial conditions

t0 = 0      # initial time
dt = 0.1    # step in time
tf = 100    # max time


# defining the equation of Hooke's Law 

def d2xdt2(x,k):
    return(-(k/m)*x)     # k = spring constant, m = mass of the object on the spring


# defining the equation for the spring function in order to test it on different values for intial velovity, displacment, mass, and spring constant.

def spring(v0,x0,k,m):
    
    # lists for the position, time, and velocity.
    x = [x0]
    t = [t0]
    v = [v0]
    
    
    # while loop using Euler equations for the position, time, and velocity
    while t[-1] < tf:
        x.append(x[-1] + v[-1]*dt)
        v.append(v[-1] + d2xdt2(x[-1],k)*dt)
        t.append(t[-1] + dt)
    return x,v,t,k,m


# Tests for the first graph, changing the intial velovity & displacment and keeping the spring constant & mass unchanged

x1, v1, t1, k1, m1 = spring(0,0,4,10)

x2, v2, t2, k2, m2 = spring(10,0,4,10)

x3, v3, t3, k3, m3 = spring(0,10,4,10)




# Tests for the second graph, changing the spring constant & mass and keeping the intial velovity & displacment unchanged

x4, v4, t4, k4, m4 = spring(10,5,1,1)

x5, v5, t5, k5, m5 = spring(10,5,50,11)

x6, v6, t6, k6, m6 = spring(10,5,7,9)

# All tests function as expected



# ploting the two graphs 

plt.figure(1)
plt.plot( t1, x1, "red")
plt.plot( t2, x2, "green")
plt.plot( t3, x3, "blue")


plt.figure(2)
plt.plot( t4, x4, "yellow")
plt.plot( t5, x5, "grey")
plt.plot( t6, x6, "black")
```
