```python
# This program is coded to describe the random motion of a particle with equal probability in 1D


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Nsteps = 1000    # number of steps
x = 0            # initial position
xlist = [x]      # a list to contain the positions of the particle


# a loop for the possible paths of the particle

for i in range(Nsteps):
    step = np.random.choice([-1,1])    # having the steps to be random with equal prabability
    x += step
    xlist.append(x)
    
    
plt.plot(xlist)
plt.xlabel("Time")
plt.ylabel("Position")
```
