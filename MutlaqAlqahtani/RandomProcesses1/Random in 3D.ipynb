```python
# this program is coded to describe the random motion of a partcle in 3D with equal probabilities along the positive and negative x, y, and z directions


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
%pylab inline

Nsteps = 1000       # number of steps
x = 0               # intial position on the x axis
y = 0               # intial position on the y axis
z = 0               # intial position on the z axis 


# list to contain the x,y,z positions
xlist = [x]
ylist = [y]
zlist = [z]

# a loop for the different paths of the particle along x , y, and z directions 
for i in range(Nsteps):
    step = np.random.choice([1,2,3,4,5,6])  # random steps with equal probabilities
    
    # conditions for the movment along the positive and negative x,y,and z directions
    
    if step == 1:
        x += 1
    elif step == 2:
        x -= 1
    elif step == 3:
        y += 1
    elif step == 4:
        y -= 1
    elif step == 5:
        z += 1
    elif step == 6:
        z -= 1
        
    xlist.append(x)
    ylist.append(y)
    zlist.append(z)

ax = plt.axes(projection='3d')

ax.plot(xlist, ylist, zlist)
```
