```python
# this program is coded to discribe the average displacment of the random motion of particles in 2D


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# defining a function to calculate the random motion

def RW():
    
    Nsteps = 100     # number of steps
    x = 0            # intial position on the x axis
    y = 0            # intial position on the y axis 
    t0 = 0           # intial time
    tf = 100         # max time
    
    
    
    # lists to contain the x and y positions
    xlist = [x]      
    ylist = [y]
    
    
    # equation and list for the displacment
    
    d = (x**2+y**2)**0.5
    dlist = [d]
    
    
    # a loop for the possible paths of the particle
      
    for i in range(Nsteps):
        step = np.random.choice([1,2,3,4])   # random steps with equal probabilities
         
        if step == 1:                        # probabilty of moving to the right
            x += 1
        elif step == 2:                      # probabilty of moving to the left
            x -= 1
        elif step == 3:                      # probabilty of moving up
            y += 1
        elif step == 4:                      # probabilty of moving down
            y -= 1
        
        d = ((x**2+y**2)**0.5)              # equation for the displacment
        xlist.append(x)
        ylist.append(y)
        dlist.append(d)
    
    
    plt.figure(1)               # the first graph for the random motion in 2D
    plt.plot(xlist, ylist)
    plt.xlabel("X")
    plt.ylabel("Y")
    
    
    
    return dlist

# creating a matrix for the displacment and trials
dd = np.zeros((500,101))

# a loop to perform the trials over the all the displacments
for i in range(500):
    dlist = RW()
    
    dd[i,:] = dlist
    
Avg = np.mean(dd, axis=0)   # finding the average of the overall displacment for all trials


plt.figure(2)               # the second graph for the average displacment
plt.plot(Avg)
```

```python
# this program is coded to discribe the average displacment of the random motion of particles in 3D

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
%pylab inline



# defining a function to calculate the random motion

def RW():
    
    Nsteps = 1000    # number of steps
    x = 0            # intial position on the x axis
    y = 0            # intial position on the y axis 
    z = 0            # intial position on the z axis
    t0 = 0           # intial time
    tf = 100         # max time
    
    
    
    # lists to contain the x, y, and z positions
    xlist = [x]      
    ylist = [y]
    zlist = [z]
    
    
    # equation and list for the displacment
    d = ((x**2+y**2+z**2)**0.5)
    dlist = [d]
    
    
    # a loop for the possible paths of the particle
      
    for i in range(Nsteps):
        step = np.random.choice([1,2,3,4,5,6])   # random steps with equal probabilities on the x,y,and z direction
         
        if step == 1:                        
            x += 1
        elif step == 2:                      
            x -= 1
        elif step == 3:                      
            y += 1
        elif step == 4:                      
            y -= 1
        elif step == 5:
            z += 1
        elif step == 6:
            z -= 1
        
        d = ((x**2+y**2+z**2)**0.5)        # equation for the displacment
        xlist.append(x)
        ylist.append(y)
        zlist.append(z)
        dlist.append(d)
    
    
    plt.figure(1)                           # the first graph for the particle in 3D
    ax = plt.axes(projection='3d')
    ax.plot(xlist, ylist, zlist)
    
    
    return dlist



# creating a matrix for the displacment and trials
dd = np.zeros((500,1001))


for i in range(500):
    dlist = RW()
    
    dd[i,:] = dlist
    
Avg = np.mean(dd, axis=0)    # finding the average of the overall displacment for all trials

plt.figure(2)
plt.plot(Avg)                # the second graph for the average displacment
```
