```python
# This program is coded to describe the spin-up and spin-down of particles with equal probabilities for spining up and down




import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# creating a matrix m to fill out with particles

m = np.zeros((100,100))

plt.figure(1)

plt.imshow(m)     # ploting a test for an empty matrix m



# creating a loop to fill out the rows and columns with spin up or down

for s in range(100):
    i = np.random.choice(100)        # filling out the rows randomly 
    j = np.random.choice(100)        # filling out the columns randomly 
    e = np.random.choice([1,-1])     # picking either -1 (spin-down) or 1 (spin-up)
    m[i,j] = e


plt.figure(2)
plt.imshow(m)       # ploting a test for only 100 steps 



# creating the same loop with more steps to cover all of the matrix

for s in range(100000):
    i = np.random.choice(100)
    j = np.random.choice(100)
    e = np.random.choice([1,-1])
    m[i,j] = e

plt.figure(3)
plt.imshow(m)    # ploting the final test to show the covered matrix


```
