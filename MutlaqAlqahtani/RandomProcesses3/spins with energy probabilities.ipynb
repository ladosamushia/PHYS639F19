```python
# This program is coded to describe the spin-up and spin-down of particles with different probabilities for spining up and down




import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# creating a matrix m to fill out with particles

m = np.zeros((100,100))

plt.figure(1)

plt.imshow(m)     # ploting a test for an empty matrix m



# creating a loop to fill out the rows and columns with spin up or down

for s in range(100000):
    i = np.random.choice(100)        # filling out the rows randomly 
    j = np.random.choice(100)        # filling out the columns randomly 
    
    # picking either -1 (spin-down) or 1 (spin-up) with different probabilities
    e = np.random.choice([1,-1], p = [0.8,0.2])
    m[i,j] = e


plt.figure(2)
plt.imshow(m)       # ploting a test for the first set of probabilites 



# creating the same loop with the probabilities reversed 

for s in range(100000):
    i = np.random.choice(100)
    j = np.random.choice(100)
    e = np.random.choice([1,-1], p = [0.2,0.8])
    m[i,j] = e

plt.figure(3)
plt.imshow(m)    # ploting the second set of probabilites to show the contradictions

```

```python
# this program is coded to describe the spin-up and spin-down of particles using the energy probabilities



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


t0 = 0            # intial time
tf = 1000         # max time
Nsteps = 100000   # number of steps

T = 273           # the tempreture


# creating an empty m matrix to be filled out

m = np.zeros((101,101))
plt.figure(1)
plt.imshow(m)         # ploting a test for an empty matrix 



# creating a loop to fill out with particles spining up or down

for s in range(Nsteps):
    i = np.random.choice(100)        # filling out the rows randomly 
    j = np.random.choice(100)        # filling out the columns randomly
    
    
    # the energy of spining-up based on the spins of nearby particles 
    Eup = -1*(m[i+1,j] + m[i-1,j] + m[i,j+1] + m[i,j-1])
    
    # the enegry of spining-down which is the opposite of the Eup
    Edown = -Eup
    
    # the probability of spining up and down using the equation for the energy 
    Pup = (np.e**(-Eup/T))/(np.e**(-Eup/T)+np.e**(-Edown/T))
    Pdown = (np.e**(-Edown/T))/(np.e**(-Eup/T)+np.e**(-Edown/T))
    
    # filling out the matrix with spin-up (1) and spin-down (-1) with energy probability
    e = np.random.choice([1,-1], p=[Pup, Pdown])   
    m[i,j] = e



plt.figure(2)     # ploting the final plot with the energy probability
plt.imshow(m)


```
