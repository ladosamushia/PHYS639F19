```python
# This code solves a first order ODE



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#intial conditions

x0=0
f0=0

# final value of x

xf = 10

# step size

dx=0.00001

# the function df/dx

def dfdx(x):
    return(np.cos(x))

# lists of values for x and the function

x=[x0]
fx=[f0]


# creating a loop to maintain the limit and construct the function

while x[-1] < xf:
    fx.append(fx[-1]+dfdx(x[-1])*dx)
    x.append(x[-1]+dx)
    
    
# Test
t= np.linspace(0,10,1000)
ft=np.sin(t)

# Ploting sin(t) for comparision 
plt.plot(t,ft,'red')

# Ploting the function
plt.plot(x,fx,'green')
```
